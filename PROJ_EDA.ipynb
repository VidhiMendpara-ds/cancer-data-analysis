```python
#RANDOM FOREST FOR TARGET SEVERITY SCORE

#Converting categprical values in numec

cat_cols = ['Gender','Country_Region', 'Cancer_Type', 'Cancer_Stage']

for col in cat_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# print(data)

#Preparong Features and Input:

X = data.drop(columns = ['Patient_ID', 'Treatment_Cost_USD', 'Survival_Years', 'Target_Severity_Score'])
y_severity = data['Target_Severity_Score']

#Train and test splitting:

X_train, X_test, y_train_s, y_test_s = train_test_split(X, y_severity, test_size = 0.2, random_state = 42)

#Train the model:

model = RandomForestRegressor(n_estimators = 200, max_depth = None, min_samples_split = 2, min_samples_leaf = 1)
model.fit(X_train, y_train_s)

#Evaluate the model:

train_r2_score = r2_score(y_train_s, model.predict(X_train))
test_r2_score = r2_score(y_test_s, model.predict(X_test))
```

```python
#RANDOM FOREST FOR TARGET SURVIVEAL YEARS

#Converting categprical values in numec

cat_cols = ['Gender','Country_Region', 'Cancer_Type', 'Cancer_Stage']

for col in cat_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# print(data)

#Preparong Features and Input:

X = data.drop(columns = ['Patient_ID', 'Treatment_Cost_USD', 'Survival_Years', 'Target_Severity_Score'])
y_sy = data['Survival_Years']

#Train and test splitting:

X_train, X_test, y_train_s, y_test_s = train_test_split(X, y_sy, test_size = 0.2, random_state = 42)

#Train the model:

param_grid = {
    'n_estimators' : [100, 200],
    'max_depth' : [5, 10, None],
    'min_samples_split' : [2,5],
    'min_samples_leaf' : [1,2]
    }

model = RandomForestRegressor(random_state = 42)
GCV = GridSearchCV(model, param_grid, cv = 3, scoring  = 'r2', n_jobs = -1)
GCV.fit(X_train, y_train_s)

#Evaluate the model:

best_rf_severity= GCV.best_estimator_

train_r2_score = r2_score(y_train_s, best_rf_severity.predict(X_train))
test_r2_score = r2_score(y_test_s, best_rf_severity.predict(X_test))

print('train_r2_score of Survival year: ', train_r2_score)
print('test_r2_score of Survival year: ', test_r2_score)
```

This highlights a clear global inequality in healthcare affordability that can intensify financial strain depending on a patients country of residence.

Gender-Based Cost Patterns Are Uniform
Across all countries, gender-based differences in average treatment costs are minimal, suggesting no major gender bias in pricing or access to cancer care. This uniformity may reflect standardization in treatment protocols or equitable healthcare policies, but it also points to the fact that the financial impact of cancer is universal across genders.

Age-Related Escalation in Treatment Costs
Treatment costs tend to rise with age, particularly for those aged 61 and above. This trend is especially evident in countries like Australia and the USA, where older age groups face sharply higher costs. The increased financial burden in these groups could be due to more intensive care needs, multiple comorbidities, or prolonged treatments. This pattern underlines the vulnerability of elderly populations and the pressing need for targeted support for senior citizens.

Role of Healthcare Systems in Cost Variation
Countries with robust public healthcare systemssuch as Canada, Germany, and the UKshow relatively stable treatment costs across age groups, reflecting the benefits of healthcare subsidies or coverage. This consistency reinforces the importance of government intervention and universal healthcare in mitigating financial disparities in cancer treatment.

```python
from scipy.stats import pearsonr, spearmanr

#importing earsonr, spearmanr cuz scipy one gives p_values too

x = data["Treatment_Cost_USD"]
y = data["Survival_Years"]

#Performing Pearson Correlation:

pearson_corr, pearson_p = pearsonr(x,y)
print(f"Pearson Correlation Coefficient: {pearson_corr}")
print(f"Pearson P-value: {pearson_p}")


spearman_corr, spearman_p = spearmanr(x,y)
print(f"spearman Correlation Coefficient: {spearman_corr}")
print(f"spearman P-value: {spearman_p}")

alpha = 0.05

def interpret_corr(corr, p, method):

    if alpha>p:
        print(f"{method}, We reject the null hypothesis")

    else:
        print(f"{method}, We failed to reject the null hypothesis")

interpret_corr(pearson_corr, pearson_p, "pearson")
interpret_corr(spearman_corr, spearman_p, "spearman")

        

```

```python
data.columns
```

```python
df = data.copy()
```

```python
stage_list = ['Stage 0', 'Stage I', 'Stage II', 'Stage III', 'Stage IV']
```

```python
grouped_stats = df.groupby('Cancer_Stage')[["Treatment_Cost_USD", "Survival_Years"]].mean().reset_index()
grouped_stats
```

```python
#FOR ANOVA:

grouped_cost = []
grouped_survival = []

for stage in stage_list:
    stage_data = df[df["Cancer_Stage"] == stage]
    cost = stage_data['Treatment_Cost_USD']
    survival = stage_data['Survival_Years']
    grouped_cost.append(cost)
    grouped_survival.append(survival)
```

```python
len(grouped_cost)
```

```python
# check for normaility by shapiro test (CUZ FOR ANNOVE IT'S MUST TO CHECK IF DATA HAS BEEN NORMALLY DESTRIBUTED AMONG EACH GROUP)
from scipy.stats import  shapiro, f_oneway
normal_cost=0
normal_survival=0

for i in range (len(stage_list)):
    cost_p= shapiro(grouped_cost[i]).pvalue
    surv_p= shapiro(grouped_survival[i]).pvalue 
    print(f" cost {cost_p} for group {i}")
    print(f"Survival {surv_p} for group {i}")
    if cost_p<0.05:
        normal_cost+=1
    if surv_p<0.05:
        normal_survival+=1

print(cost_p)
print(surv_p)
```
